```python
import pandas as pd
from pandas.plotting import autocorrelation_plot
import matplotlib.pyplot as plt
import datetime
from pdb import set_trace
```

```python
logistics_path = '../data/PrimaryData/contraceptive_logistics_data.csv'
df = pd.read_csv(logistics_path)

# Create date column from year/month columns
df['date'] = df['month'].map(str)+ '-' +df['year'].map(str)
df['date'] = pd.to_datetime(df['date'], format='%m-%Y').dt.to_period('M')

# Reorder to put date first. Remove year and month columns
df_columns = df.columns.to_list()
col_to_remove = ['date','year','month']
df_columns = [col for col in df_columns if col not in col_to_remove]
df_columns = ['date'] + df_columns
df = df[df_columns]

df.sort_values(by=['date'], inplace=True)
df.reset_index(inplace=True, drop=True)
```

```python
cols_to_print = ['region','district','product_code','site_code']
for col in cols_to_print:
    print('Number {}: {}'.format(col, len(df[col].unique())))
    print(df[col].unique())
```

```python
# Verifying no duplicate rows oiverall
dups_cols = ['date','region','district','product_code','site_code']
dups = df.duplicated(subset=dups_cols, keep=False)
df[dups][dups_cols]
```

```python
# Count site codes per district
location_params = ['region','district','site_code']
site_df = df[location_params].drop_duplicates().sort_values(by=['district','region'])
site_counts = site_df.pivot(columns='district',values='site_code').count()
print('Number of site codes per district')
plt.subplot(121)
site_counts.hist()
plt.title('Site Codes per District')

# Count districts per region
location_params = ['region','district']
site_df = df[location_params].drop_duplicates().sort_values(by=['district','region'])
site_counts = site_df.pivot(columns='region',values='district').count()
print('Number of districts per region')
plt.subplot(122)
site_counts.hist()
plt.title('Districts per Region')
```

```python
# Plot overall mean and median trends
# Mean
plot_cols = ['stock_distributed']
filtered_mean = df.groupby(['date'], as_index=False).mean()
ax = filtered_mean.plot(title='Mean Trend', x='date', y=plot_cols, kind='line', figsize=(15,10))
mean = df['stock_distributed'].mean()
plt.plot(filtered_mean['date'], [mean for count in range(len(filtered_mean))],'--r')

# Median
filtered_median = df.groupby(['date'], as_index=False).median()
filtered_median.plot(ax=ax, title='Over trends', x='date', y=plot_cols, kind='line', figsize=(15,10), color='g')
ax.legend(["Mean Stock Distributed", "Overall Mean Stock Distributed Per Month", "Median Stock Distributed"])
```

```python
ax = autocorrelation_plot(filtered_mean['stock_distributed'])
ax.set_title('Autocorrelation of Mean Stock Distributed')
ax.set_xlabel('Lag (Months)')
```
